```python
from konlpy.tag import Kkma
from konlpy.corpus import kolaw
from konlpy.utils import pprint
from nltk import collocations

from collections import Counter

import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud


# read data 
data1 = pd.read_csv('yanolja.csv')
data2 = pd.read_csv('woowahan.csv')
data3 = pd.read_csv('univ.csv')
data4 = pd.read_csv('zigbang.csv')
data5 = pd.read_csv('coupang.csv')

# merge the data of startups 
data = pd.concat([data1,data2, data3, data4, data5], ignore_index = True)
```

```python
# get the bigram of merits
measures = collocations.BigramAssocMeasures()
doc = str(list(data['merit']))

#print('\nCollocations among tagged words:')
tagged_words = Kkma().pos(doc)
finder = collocations.BigramCollocationFinder.from_words(tagged_words)
#pprint(finder.nbest(measures.pmi, 50)) # top 5 n-grams with highest PMI

print('\nCollocations among words:')
words = [w for w, t in tagged_words]
finder = collocations.BigramCollocationFinder.from_words(words)
finder.apply_word_filter(lambda w: len(w) < 2)
finder.apply_freq_filter(3) # only bigrams that appear 3+ times
good_scored = finder.score_ngrams(measures.pmi)
```

```python
# get the bigram of disadvantages 
measures = collocations.BigramAssocMeasures()
doc = str(list(data['disadvantage']))

#print('\nCollocations among tagged words:')
tagged_words = Kkma().pos(doc)
finder = collocations.BigramCollocationFinder.from_words(tagged_words)
#pprint(finder.nbest(measures.pmi, 50)) # top 5 n-grams with highest PMI

print('\nCollocations among words:')
words = [w for w, t in tagged_words]
finder = collocations.BigramCollocationFinder.from_words(words)
finder.apply_word_filter(lambda w: len(w) < 2)
finder.apply_freq_filter(3) # only bigrams that appear 3+ times
bad_scored = finder.score_ngrams(measures.pmi)
```

```python
raw_bigram_good = dict()
raw_bigram_bad = dict()

for i in range(len(good_scored)):
    bigram = good_scored[i][0][0] + good_scored[i][0][1]
    count = 0
    for j in range(len(data['merit'])):
        if count == 1000:
            break
        if bigram in data['merit'][j].replace(" ", ""):
            print(data['merit'][j] + "\n")
            count += 1
            
    if count >= 1:
        raw_bigram_good[bigram] = count
            

for i in range(len(bad_scored)):
    bigram = bad_scored[i][0][0] + bad_scored[i][0][1]
    count = 0
    for j in range(len(data['disadvantage'])):
        if count == 1000:
            break
        if bigram in data['disadvantage'][j].replace(" ", ""):
            print(data['disadvantage'][j] + "\n")
            count += 1
            
    if count >= 1:
        raw_bigram_bad[bigram] = count
            
```

```python
# wordcloud hue for disadvantage (red)
def random_color_func(word=None, font_size=None, position=None,  orientation=None, font_path=None, random_state=None):
    h = int(360.0 * 153.0 / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)

    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
# wordcloud hue for merit (blue)
def random_color_func2(word=None, font_size=None, position=None,  orientation=None, font_path=None, random_state=None):
    h = int(360.0 * 10.0 / 255.0)
    s = int(100.0 * 255.0 / 255.0)
    l = int(100.0 * float(random_state.randint(60, 120)) / 255.0)

    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
good_cloud2 = WordCloud(width=900, height=600, font_path = r"‪C:\Windows\Fonts\GodoB.ttf", background_color='white', color_func=random_color_func)

good_cloud2 = good_cloud2.fit_words(raw_bigram_good)

plt.figure(figsize=(15,20))
plt.axis('off')
plt.imshow(good_cloud2)
plt.show()

bad_cloud2 = WordCloud(width=900, height=600, font_path = r"‪‪‪C:\Windows\Fonts\GodoB.ttf", background_color='white', color_func=random_color_func2)

bad_cloud2 = bad_cloud2.fit_words(raw_bigram_bad)

plt.figure(figsize=(15,20))
plt.axis('off')
plt.imshow(bad_cloud2)
plt.show()
```
